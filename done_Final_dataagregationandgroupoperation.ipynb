```python
#Quantile and bucket analysis
frame = DataFrame({'data1': np.random.randn(10),
                   'data2': np.random.randn(10)})
frame
```

```python
factor=pd.cut(frame.data1,4)
factor[:10]
```

```python
def get_stats(group):
    return {'min':group.min(),'max':group.max(),
            'count': group.count(), 'mean':group.mean()}
grouped = frame.data2.groupby(factor) #frame.data2 selects data2 column of dataframe named frame and factor puts each value as per the defined bucket
print (grouped)
grouped.apply(get_stats).unstack()
```

```python
#The above were equal length buckets to compute equal size bucket use quantile
grouping =pd.qcut(frame.data1, 10, labels=False)
grouped = frame.data2.groupby(grouping)
grouped.apply(get_stats).unstack()
```

```python
#replacing nan values with group mean
states = ['Ohio','New York','Vermont','Florida','Oregon','Nevada','California','Idaho']
group_key = ['East']*4 +['West']*4
data =Series(np.random.randn(8), index= states)
data[['Vermont','Nevada','Idaho']]=np.nan
data.groupby(group_key).mean()
```

```python
#we can fill the nan values with these means on the basis of whether they are East or West
fill_mean = lambda g: g.fillna(g.mean())
data.groupby(group_key).apply(fill_mean)
```

```python
# We can also substitute the NAN values with predefined values
fill_values ={'East':0.5, 'West': -1}
fill_func = lambda g :g.fillna(fill_values[g.name])
data.groupby(group_key).apply(fill_func)
```

```python
#Group weighted average and corelation
df=DataFrame({'category':['a','a','a','a','b','b','b','b'],
             'data':np.random.randn(8),'weights':np.random.randn(8)})
df
```

```python
grouped=df.groupby('category')
get_wavg = lambda g: np.average(g['data'],weights=g['weights'])
grouped.apply(get_wavg)
```

```python
#Group wise linear regression
import statsmodels.api as sm
def regress (data, yvar, xvars):
    Y = data[yvar]
    X = data[xvars]
    X['intercept']=1.
    result =sm.OLS(Y,X).fit()
    return result.params
```

```python
by_year.apply(regress, 'AAPL', ['SPX'])
```
