```python
#Regular expression ,commonly called regex
#whitespace characters are tab, space, newline
#the regex describing one or more whitespace is \s+
import re
text ="foo       bar\t bax      \tqux"
re.split('\s+', text)
```

```python
regex=re.compile('\s+')
print (regex)
regex.split(text)
```

```python
regex.findall(text) # regex include whitespace character and findall will return the matching string i.e. whitespace in this case
```

```python
#findall returns all matches, search return only the first match, match only match the beginning of the string
text="""Dave dave@google.com
Steve steve@gmail.com
Rob rob@gmail.com
Ryan ryan@yahoo.com
"""
pattern =r'[A-Z0-9._%+-]+@[A-Z0-9.-]+\.[A-Z]{2,4}'
#re.IGNORECASE makes the regex case insentitive
regex=re.compile(pattern,flags=re.IGNORECASE)
regex.findall(text)
```

```python
#search
m=regex.search(text)
m
text[m.start():m.end()] #will return the first object that is matched
#match only if pattern occurs at the start of string
print (regex.match(text))
```

```python
# sub will return new string with occurences of the pattern replaced by the new string
print (regex.sub('REDACTED', text))
```
